```python
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
high_line = go.Bar(
                    x = high20.country,
                    y = high20.literacy,
                    name = "Large Countries",
                    marker = dict(color = 'blue'),
                    text= high20.country)
```

```python
low_line = go.Bar(
                    x = low20.country,
                    y = low20.literacy,
                    name = "Small Countries",
                    marker = dict(color = 'red'),
                    text= low20.country)
```

```python
compare = [high_line, low_line]
layout = dict(title = 'Large vs Small Countries in Literacy %',
              xaxis= dict(ticklen= 5,zeroline= True))
fig = dict(data = compare, layout = layout)
py.iplot(fig)
```
